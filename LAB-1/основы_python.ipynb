#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here
def count_vowels(s):
    vowels = set("aeiouAEIOU")

    count = sum(1 for char in s if char in vowels)

    return count
```

```python
print(count_vowels("abcdefg"))
print(count_vowels("hello world"))
print(count_vowels("machine learning"))
print(count_vowels("alper guzel"))
print(count_vowels("xyz"))
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    step=0
    while n >=10:
        multiplication=1
        for digit in str(n):
            multiplication *= int(digit)
        n = multiplication
        step += 1
    return step
```

```python
def test_magic():
    print(magic(39))   # Output: 3
    print(magic(999))  # Output: 4
    print(magic(4))    # Output: 0
    print(magic(25))   # Output: 2
    print(magic(77))   # Output: 4

    print("All test cases passed!")
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import numpy as np                            

def mse(pred, true):                               
    assert len(pred) == len(true), "Vektör uzunlukları eşit olmalı!"  

    pred = np.array(pred)                           
    true = np.array(true)                           

    differences = pred - true                      
    squared = differences ** 2
    mean_squared_error = np.mean(squared)           

    return mean_squared_error                       
```

```python
def test_mse():                                        
    assert np.isclose(mse([1,2,3], [1,2,3]), 0.0)       
    assert np.isclose(mse([2,3,4], [1,2,3]), 1.0)       
    assert np.isclose(mse([10,20,30], [10,25,40]), 41.6666666667)  
    print("успешно")

test_mse() 
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):                            
    i = 2                                        
    factors = []                                  

    while i * i <= n:                             
        count = 0
        while n % i == 0:                         
            n //= i                               
            count += 1                            
        if count > 0:                             
            factors.append((i, count))             
        i += 1                                    
    if n > 1:                                     
        factors.append((n, 1))                    

    # формируем строку результата
    result = ""                                   
    for (p, k) in factors:                        
        if k == 1:                                
            result += f"({p})"                    
        else:
            result += f"({p}**{k})"               
    return result                                
```

```python
print(prime_factors(86240))   # (2**5)(5)(7**2)(11)
print(prime_factors(60))      # (2**2)(3)(5)
print(prime_factors(13))      # (13)
print(prime_factors(360))     # (2**3)(3**2)(5)
```

```python
def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(60) == "(2**2)(3)(5)"
    assert prime_factors(13) == "(13)"
    assert prime_factors(360) == "(2**3)(3**2)(5)"
    assert prime_factors(1) == ""                      
    print("✅  успешно")

test_prime_factors()

```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    a= 0

    while total < number:
       
        a += 1
        total += a**2
        if total == number:
            return a
        
    return "nivazmojna"
        
        
   
   
```

```python
print(pyramid(1))      # 1 = 1² → k = 1
print(pyramid(5))      # 1² + 2² = 5 → k = 2
print(pyramid(14))     # 1² + 2² + 3² = 14 → k = 3
print(pyramid(30))     # 1² + 2² + 3² + 4² = 30 → k = 4
print(pyramid(7))      # невозможно → "It is impossible"
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def balance(number):
    s = str(number)
    length = len(s)
    if length == 1:
        return True

    if length % 2 == 0:                 
        mid1 = length // 2 - 1          
        mid2 = length // 2              
        left = s[:mid1]                 
        right = s[mid2 + 1:]            
    else:                               
        mid = length // 2
        left = s[:mid]                  
        right = s[mid + 1:]

    sum_left = sum(int(x) for x in left)
    sum_right = sum(int(x) for x in right)

    return sum_left == sum_right


```

```python
def test_balance():
    assert balance(1238033) == True     
    assert balance(123321) == True       
    assert balance(1234) == False        
    assert balance(5) == True           
    assert balance(41302) == False       
    assert balance(1114111) == True      
    print(" успешно !")

test_balance()
```
