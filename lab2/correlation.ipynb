## Часть 3: Вычисление корреляции с Python


#### Шаг 1: Рассчитайте корреляцию для brainFrame.
Метод `corr()` библиотеки pandas обеспечивает простой способ вычисления корреляции для кадра данных. Просто применяя метод на кадре данных, можно получить корреляцию между всеми переменными одновременно.

```python

numeric_columns = brainFrame.select_dtypes(include=['float64', 'int64']).columns
brainFrame[numeric_columns].corr(method='pearson')
```

Обратите внимание на диагональ слева направо в таблице корреляции, сгенерированной выше. Почему диагональ заполнена значениями 1? Это совпадение? Объясните.

Это не совпадение. Значения 1 на диагонали всегда присутствуют в корреляционной матрице, потому что они показывают корреляцию переменной с самой собой. Корреляция переменной с самой собой всегда равна 1 (идеальная положительная корреляция), так как переменная изменяется точно так же, как и она сама.

Продолжая смотреть на таблицу корреляции выше, обратите внимание, что значения зеркалируются; значения под диагональю имеют зеркальный аналог над ней. Это совпадение? Объясните.

Это тоже не совпадение. Зеркальное отображение значений является математическим свойством корреляции, потому что корреляция между переменными A и B всегда равна корреляции между B и A. Например, если корреляция между ростом и весом равна 0.7, то и корреляция между весом и ростом тоже будет 0.7. Это свойство называется симметричностью корреляции и является фундаментальным свойством корреляционной матрицы.

Используя тот же метод `corr()`, вычислите корреляцию переменных, содержащихся в кадре с женскими записями:

```python

numeric_columns = womenDf.select_dtypes(include=['float64', 'int64']).columns
womenDf[numeric_columns].corr(method='pearson')
```

И то же самое можно сделать для переменных, содержащихся в кадре с мужскими записями:

```python

numeric_columns = menDf.select_dtypes(include=['float64', 'int64']).columns
menDf[numeric_columns].corr(method='pearson')
```

#### Шаг 2: Нарисуйте корреляционную тепловую карту.

Теперь, когда кадры данных готовы, можно отобразить тепловые карты. Далее приведен анализ кода из ячейки ниже:

Строка 1: Создает таблицу корреляции, основанную на фрейме данных `womenNoGenderDf` и хранит его в `wcorr`.<br>
Строка 2: Использует метод `heatmap()` библиотеки `seaborn` для генерации и построения тепловой карты. Обратите внимание, что `heatmap()` принимает `wcorr`, как параметр.<br>
Строка 3: используется для экспорта и сохранения созданной тепловой карты в виде PNG-изображения. Хотя строка 3 не активна (перед ней стоит символ `#`, заставляя интерпретатор игнорировать ее), она сохранена в информационных целях.

```python
# Ячейка для кода № 12
import seaborn as sns


num = womenDf.select_dtypes(include=['number'])
wcorr = num.corr()


sns.heatmap(wcorr, annot=True, cmap='coolwarm')
plt.tight_layout()
plt.savefig('attribute_correlations.png')
plt.show()
```

Аналогичным образом создайте и отобразите тепловую карту для кадра данных с мужчинами.

```python
# Ячейка для кода № 14
num = menDf.select_dtypes(include=['number'])
mcorr = num.corr()

sns.heatmap(mcorr, annot=True, cmap='coolwarm')
plt.tight_layout()
plt.savefig('attribute_correlations.png')
plt.show()
#
#
```

У многих пар переменных корреляция близка к нулю. Что это значит?

 это означает отсутствие заметной линейной связи между этими переменными — изменение одной почти не предсказывает изменение другой. Это не исключает нелинейной зависимости, влияния выбросов/шума или скрытых подгрупп (например, по полу). Проверьте scatter‑plot и Spearman для дополнительной проверки.

Зачем делать разделение по полу?

Чтобы избежать искажения результатов: пол может быть конфаундером и влиять на измеряемые переменные, скрывая или изменяя связи. Разделение по полу позволяет посмотреть корреляции внутри каждой группы отдельно и обнаружить скрытые зависимости.
